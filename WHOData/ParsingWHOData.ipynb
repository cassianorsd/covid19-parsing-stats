```python
import urllib.request
from tabula import read_pdf
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
pdfURL = 'https://www.who.int/docs/default-source/coronaviruse/situation-reports/20200329-sitrep-69-covid-19.pdf?sfvrsn=8d6620fa_2'
urllib.request.urlretrieve(pdfURL,'./29-03-2020.pdf')
```

```python
res = read_pdf('./29-03-2020.pdf',pages=[2,3,4,5,6])
```

```python
dfs = []
for df in res:
    headers = ['Country','Confirmed','NewCases','Deaths','NewDeaths']
    oldHeaders = df.columns.values
    df = df[df.columns[0:5]]
    df.columns = headers
    df = df.append(pd.DataFrame([oldHeaders[0:5]],columns=headers))
    dfs.append(df)
```

```python
df = pd.concat(dfs)
df
```

```python
df['Confirmed'] = pd.to_numeric(df.Confirmed,errors='coerce')
df['NewCases'] = pd.to_numeric(df.NewCases,errors='coerce')
df['Deaths'] = pd.to_numeric(df.Deaths,errors='coerce')
df['NewDeaths'] = pd.to_numeric(df.NewDeaths,errors='coerce')
df.dropna(subset=['Country','Confirmed'],inplace=True)
df = df[df['Country']!='Subtotal for all regions']
df = df[df['Country']!='Grand total']
```

```python
df[df.Country=='Brazil']
```

```python
plt.figure(figsize=(10,5))
ax = sns.barplot(x='Country',y='Confirmed',palette="Set2",data=df.sort_values(by='Confirmed',ascending=False).head(20))
ax.set_xticklabels(ax.get_xticklabels(), rotation=90)
plt.show()
```

```python
plt.figure(figsize=(10,5))
ax = sns.barplot(x='Country',y='Deaths',palette="Set1",data=df.sort_values(by='Deaths',ascending=False).head(20))
ax.set_xticklabels(ax.get_xticklabels(), rotation=90)
plt.show()
```

```python
plt.figure(figsize=(10,5))
ax = sns.barplot(x='Country',y='NewCases',palette="Set2",data=df.sort_values(by='NewCases',ascending=False).head(20))
ax.set_xticklabels(ax.get_xticklabels(), rotation=90)
plt.show()
```
